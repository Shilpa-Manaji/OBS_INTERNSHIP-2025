```python
def convert_currency(value):
    value=str(value).replace("€","")
    if "K" in value:
        value=float(value.replace("K" ,""))*1000
    elif "M" in value:
        value=float(value.replace("M" ,""))*1000000
    elif value==np.nan:
        value=np.nan
    else:
        value=float(value)
    return value                      
```

```python
#converting it into M 

df["Value"] = df["Value"].apply(convert_currency)/100000
df["Wage"] =  df["Wage"].apply(convert_currency)/1000000           
df["Release Clause"]=df["Release Clause"].apply(convert_currency)/1000000
```

### Detect and Handle Outliers

```python
# To detect Outliers 
def detect_outliers(column):
    Q1 = df[column].quantile(0.25)  # First quartile
    Q3 = df[column].quantile(0.75)  # Third quartile
    Q2 = Q3 - Q1                    # Interquartile Range

    lower_bound = Q1 - 1.5 * Q2 
    upper_bound = Q3 + 1.5 * Q2

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Check for outliers in important numerical columns
outlier_columns = ["Overall", "Potential", "Value", "Wage","Release Clause"]
for col in outlier_columns:
    print(f"Outliers in {col}:\n", detect_outliers(col))
```

```python
#Capping the outliers

def cap_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    Q2 = Q3 - Q1  

    lower_bound = Q1 - 1.5 * Q2
    upper_bound = Q3 + 1.5 * Q2

    df[column] = np.clip(df[column], lower_bound, upper_bound)

# Apply Winsorization
for col in ["Overall", "Potential", "Value", "Wage", "Release Clause"]:
    cap_outliers(col)
```
