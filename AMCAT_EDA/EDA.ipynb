```python
# Filter dataset for Computer Science graduates in relevant job roles
cs_grads = df[(df["Specialization"] == "CS") & 
    (df["Designation"].isin(["programming analyst", "software engineer", "hardware engineer", "associate engineer"]))]

# Compute Salary Statistics
salary_stats = cs_grads["Salary"].describe()

# Print results
print("Salary Summary for Fresh CSE Graduates in Given Roles:")
print(salary_stats)

# Visualize Salary Distribution
plt.figure(figsize=(10, 5))
sns.barplot(x=cs_grads['Designation'],y=cs_grads['Salary']/10000,hue=cs_grads['Specialization'])
plt.title("Salary Distribution of Fresh CSE Graduates")
plt.xlabel("Specialization")
plt.ylabel("Salary (in LPA)")
plt.show()
```

yes, **Software engineer and associate engineers** with CS specialization have **2.5 and above** Lakh per annum salary

```python
from scipy.stats import chi2_contingency

# Plot a count plot to visualize specialization preference by gender
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Specialization', hue='Gender')
plt.xticks(rotation=45)
plt.title('Specialization Preference by Gender')
plt.xlabel('Specialization')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()

# Create a contingency table for Chi-square test
contingency_table = pd.crosstab(df['Gender'], df['Specialization'])

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print results
print(f'Chi-square Statistic: {chi2}')
print(f'P-value: {p}')

# Interpretation
if p < 0.05:
    print("There is a significant relationship between Gender and Specialization preference.")
else:
    print("There is no significant relationship between Gender and Specialization preference.")
```

**Observations from Specialization Preferences by Gender**  

- **CS, EC, and IT** are the top choices for both genders, with males significantly outnumbering females.  
- **Male students dominate Mechanical (ME) and Electrical (EE) fields,** while females show lower representation.  
- **Less** common specializations like Civil, Chemical, and Telecommunication have very few students, regardless of gender.
